```python
import matplotlib.pyplot as plt
from sequences_io import CHIMPANZEE_PATH, HUMAN_PATH, RABBIT_PATH, read_sequences
from utils import NucleotideMappingMethod, SequenceRepresentationFactory
```

# Numerical Representations

```python
s1_1 = SequenceRepresentationFactory.create("ACTG", NucleotideMappingMethod.RUDIMENTARY)
s2_1 = SequenceRepresentationFactory.create(
    "ACTGA", NucleotideMappingMethod.RUDIMENTARY
)
s3_1 = SequenceRepresentationFactory.create(
    "ACTGAC", NucleotideMappingMethod.RUDIMENTARY
)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(20, 5))
ax1 = s1_1.plot_representation(ax=axs[0])
ax1.set_title("ACTG")
ax2 = s2_1.plot_representation(color="r", ax=axs[1])
ax2.set_title("ACTGA")
ax3 = s3_1.plot_representation(color="g", ax=axs[2])
ax3.set_title("ACTGAC")
plt.show()
```

```python
s1_2 = SequenceRepresentationFactory.create("AGTC", NucleotideMappingMethod.REFINED)
s2_2 = SequenceRepresentationFactory.create("AGTCA", NucleotideMappingMethod.REFINED)
s3_2 = SequenceRepresentationFactory.create("AGTCAG", NucleotideMappingMethod.REFINED)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(20, 5))
ax1 = s1_2.plot_representation(ax=axs[0])
ax1.set_title("ACTG")
ax2 = s2_2.plot_representation(color="r", ax=axs[1])
ax2.set_title("ACTGA")
ax3 = s3_2.plot_representation(color="g", ax=axs[2])
ax3.set_title("ACTGAC")
plt.show()
```

```python
s1_3 = SequenceRepresentationFactory.create("AGTC", NucleotideMappingMethod.HUFFMAN, "AGTC")
s2_3 = SequenceRepresentationFactory.create("AGTCA", NucleotideMappingMethod.HUFFMAN, "AGTCA")
s3_3 = SequenceRepresentationFactory.create("AGTCAG", NucleotideMappingMethod.HUFFMAN, "AGTCAG")
```

```python
fig, axs = plt.subplots(1, 3, figsize=(20, 5))
ax1 = s1_3.plot_representation(ax=axs[0])
ax1.set_title("ACTG")
ax2 = s2_3.plot_representation(color="r", ax=axs[1])
ax2.set_title("ACTGA")
ax3 = s3_3.plot_representation(color="g", ax=axs[2])
ax3.set_title("ACTGAC")
plt.show()
```

```python
human_seq = SequenceRepresentationFactory.create(
    read_sequences(HUMAN_PATH).seq, NucleotideMappingMethod.RUDIMENTARY
)
chimpanzee_seq = SequenceRepresentationFactory.create(
    read_sequences(CHIMPANZEE_PATH).seq, NucleotideMappingMethod.RUDIMENTARY
)
rabbit_seq = SequenceRepresentationFactory.create(
    read_sequences(RABBIT_PATH).seq, NucleotideMappingMethod.RUDIMENTARY
)

fig = plt.figure(figsize=(10, 12))
plt.title("Coding sequences of β-globin genes (rudimentary representation)")
ax1 = human_seq.plot_representation(linewidth=0.5, markersize=2)
ax2 = chimpanzee_seq.plot_representation(color="r", linewidth=1, markersize=2)
ax3 = rabbit_seq.plot_representation(color="g", linewidth=0.5, markersize=2)
ax1.legend(["Human", "Chimpanzee", "Rabbit"], loc="lower right")
plt.show()
```

```python
human_seq = SequenceRepresentationFactory.create(
    read_sequences(HUMAN_PATH).seq, NucleotideMappingMethod.REFINED
)
chimpanzee_seq = SequenceRepresentationFactory.create(
    read_sequences(CHIMPANZEE_PATH).seq, NucleotideMappingMethod.REFINED
)
rabbit_seq = SequenceRepresentationFactory.create(
    read_sequences(RABBIT_PATH).seq, NucleotideMappingMethod.REFINED
)

fig = plt.figure(figsize=(15, 7))
plt.title("Coding sequences of β-globin genes (refined representation)")
ax1 = human_seq.plot_representation(linewidth=0.5, markersize=2)
ax2 = chimpanzee_seq.plot_representation(color="r", linewidth=0.5, markersize=2)
ax3 = rabbit_seq.plot_representation(color="g", linewidth=0.5, markersize=2)
ax1.legend(["Human", "Chimpanzee", "Rabbit"], loc="lower right")
plt.show()
```

```python
human_seq = SequenceRepresentationFactory.create(
    read_sequences(HUMAN_PATH).seq,
    NucleotideMappingMethod.HUFFMAN,
    huffman_code_string=read_sequences(HUMAN_PATH).seq,
)
chimpanzee_seq = SequenceRepresentationFactory.create(
    read_sequences(CHIMPANZEE_PATH).seq,
    NucleotideMappingMethod.HUFFMAN,
    huffman_code_string=read_sequences(CHIMPANZEE_PATH).seq,
)
rabbit_seq = SequenceRepresentationFactory.create(
    read_sequences(RABBIT_PATH).seq,
    NucleotideMappingMethod.HUFFMAN,
    huffman_code_string=read_sequences(RABBIT_PATH).seq,
)

fig = plt.figure(figsize=(15, 7))
plt.title("Coding sequences of β-globin genes (huffman representation)")
ax1 = human_seq.plot_representation(linewidth=0.5, markersize=2)
ax2 = chimpanzee_seq.plot_representation(color="r", linewidth=0.5, markersize=2)
ax3 = rabbit_seq.plot_representation(color="g", linewidth=0.5, markersize=2)
ax1.legend(["Human", "Chimpanzee", "Rabbit"], loc="lower right")
plt.show()
```
